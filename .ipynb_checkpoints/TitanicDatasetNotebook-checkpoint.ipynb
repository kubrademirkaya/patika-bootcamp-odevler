## 2. Keşifsel Veri Analizi (Exploratory Data Analysis - EDA)

### 1. Kazada ölenlerin yaş ortalamasını bulunuz

### Filtreleme
#### dataset[dataset['Survived'] == 0] (Zincirleme İndeksleme - Chained Indexing)

Bu yöntem, aslında iki ayrı işlem yapar:

1. dataset['Survived'] == 0: Bu kısım, 'Survived' sütununda değeri 0 olan satırlar için True, diğerleri için False değerleri içeren bir boolean serisi oluşturur.

2. dataset[...]: Bu boolean serisi, orijinal dataset'i filtrelemek için kullanılır. Yani, True olan satırlar seçilir.

##### Düşün: Önce bir listeden belirli özelliklere sahip olanların bir listesini çıkarıyorsun, sonra o listeden tekrar seçim yapıyorsun gibi.

#### dataset.loc[dataset['Survived'] == 0]

Bu yöntem, tek bir işlem olarak kabul edilir. .loc bir indeksleme aracıdır ve ona satırları seçmek için bir koşul (boolean serisi) veriyoruz.

##### Düşün: Doğrudan listeye gidip, "sadece şu özelliklere sahip olanları al" diyorsun gibi.

```python
dead = dataset[dataset['Survived'] == 0]
dead = dataset.loc[dataset['Survived'] == 0] 
# iki satır da aynı sonucu vermektedir. Fakat .loc ile yapılan yöntem daha performanslı çalışan tercih edilen yöntemdir.
```

# Temel Dağılımlar ve İlişkiler

## Sayısal Değişkenlerin Dağılımı

### Histogram Grafik ve Yorumlaması

- Matplotlib -> hist()
- Seaborn -> histplot()

Histogramların genel yorumlamasını Fare Distribution grafiği üzerinden inceleyelim.

##### Dağılımın Genel Şekli (Sağa Çarpık, Sola Çarpık, Simetrik)

Bu grafiğin sağa çarpık (Right-Skewed) olduğunu gözlemliyoruz. Histogramın büyük bir kısmı düşük bilet fiyatları aralığında yoğunlaşmıştır ve daha yüksek bilet fiyatlarına doğru uzayan uzun bir kuyruk vardır. Bu, çoğu yolcunun nispeten ucuz biletler aldığını, ancak az sayıda yolcunun çok daha pahalı biletler aldığını gösterir (örneğin, 1. sınıf yolcuları).

##### Tepe Noktalarının Sayısı (Tek Tepe, Çift Tepe, vb.)

Bu grafikte tek tepe (unimodal) görülmektedir. Tek bir belirgin tepe noktası (mod) düşük bilet fiyatları aralığında görülür. Bu, en sık görülen bilet fiyatı aralığını temsil eder.

##### Aykırı Değerler (Outliers)

Histogramın sağ kuyruğunda, diğer fiyatlardan çok daha yüksek değerlere sahip birkaç çubuk veya yoğunluk eğrisinde ayrı bir "omuz" görülebilir. Bunlar, çok pahalı biletleri temsil eden potansiyel aykırı değerlerdir.

```python
ages = dataset["Age"]
```

```python
# Matplotlib ile en basit şekilde histogram çizimi
plt.hist(ages)
plt.show()
```

```python
# Seaborn ile en basit şekilde histogram çizimi
sns.histplot(ages)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(ages, bins=30, edgecolor='black', alpha=0.7)
plt.title('Age Distribution of Passengers (Matplotlib)')
plt.xlabel('Age')
plt.ylabel('Passenger Count')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ages, bins=30, kde=True, edgecolor='black', alpha=0.7) # kde=False yoğunluk eğrisini kapatır
plt.title('Age Distribution of Passengers (Seaborn)')
plt.xlabel('Age')
plt.ylabel('Passenger Count')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
fares = dataset['Fare']
```

```python
plt.figure(figsize = (10,6))
sns.histplot(fares, kde=True, bins= 50, edgecolor = 'black', alpha = 0.7)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Passenger Count')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

### Yoğunluk Grafiği ve Yorumlaması

- Seaborn -> kdeplot()

Elde ettiğimiz KDE grafiği, dağılımın genel şeklini daha yumuşak bir eğriyle temsil eder. Histogramdaki gibi keskin çubuklar yerine, yoğunluğun nerede daha fazla veya daha az olduğunu sürekli bir şekilde görebiliriz.

##### Tepe Noktaları (Peaks)

Eğrideki en yüksek noktalar, yoğunluğun en fazla olduğu aralıkları gösterir. Birden fazla tepe noktası (multimodal dağılım) varsa, bu veri setinde farklı grupların yoğunlaştığını gösterebilir.

##### Eğrinin Genişliği

Eğrinin ne kadar geniş olduğu, x eksenindeki verinin ne kadar yayıldığını gösterir. Age Density Distribution of Passengers grafiğini göz önüne alırsak geniş bir eğri, yaşların geniş bir aralığa yayıldığı anlamına gelirken, dar bir eğri yaşların belirli bir aralıkta toplandığını gösterir.

##### Çarpıklık (Skewness)

Eğrinin sağa veya sola doğru uzun bir kuyruğu olması, dağılımın çarpık olduğunu gösterir (tıpkı histogramda olduğu gibi).


```python
plt.figure(figsize=(10,6))
sns.kdeplot(ages, fill=True, color="skyblue", alpha=0.7)
plt.title('Age Density Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.kdeplot(fares, fill=True, color ='orchid', alpha = 0.7)
plt.title('Fare Density Distribution of Passengers')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

### Kutu Grafiği ve Yorumlaması

- Seaborn -> boxplot()

Bir kutu grafiği aşağıdaki temel bileşenleri içerir:

##### Kutu (Box): 

Verinin %25'lik (alt çeyrek, Q1) ile %75'lik (üst çeyrek, Q3) aralığını temsil eder. Kutunun içindeki çizgi medyanı (ortanca değeri) gösterir.

##### Bıyıklar (Whiskers):

Kutunun her iki ucundan uzayan çizgilerdir. Genellikle verinin yayılımının bir ölçüsünü gösterir. Varsayılan olarak, Seaborn bıyıkları Q1 - 1.5 * IQR ve Q3 + 1.5 * IQR değerlerine kadar uzatır (IQR, interquartile range = Q3 - Q1).

##### Aykırı Değerler (Outliers):

Bıyıkların ötesinde kalan noktalar, potansiyel aykırı değerleri temsil eder. Bunlar, veri setinin geri kalanından önemli ölçüde farklı olan değerlerdir.

```python
plt.figure(figsize = (8,6))
sns.boxplot(ages, color='mediumorchid')
plt.title('Age Distribution of Passengers (Box Graph)')
plt.ylabel('Ages')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

Age Distribution of Passengers grafiği üzerinden inceleme yaparsak;

Kutu (Box): Mor renkli kutu, yolcuların yaşlarının %50'sinin (yani ikinci ve üçüncü çeyrekler arasındaki değerlerin) bulunduğu aralığı temsil eder. Kutunun alt sınırı yaklaşık olarak 20 yaş civarında, üst sınırı ise yaklaşık olarak 38 yaş civarında görünüyor. Bu, yolcuların yarısının bu yaş aralığında olduğunu gösterir.

Medyan (Orta Çizgi): Kutunun içindeki yatay çizgi medyan yaşı temsil eder. Bu çizgi yaklaşık olarak 28 yaş civarında bulunuyor. Bu, yolcuların yarısının 28 yaşından küçük, diğer yarısının ise 28 yaşından büyük olduğu anlamına gelir.

Bıyıklar (Whiskers): Kutunun altından ve üstünden uzayan ince çizgiler bıyıkları temsil eder. Alt bıyık 0 yaşına kadar uzanırken, üst bıyık yaklaşık 64 yaşına kadar uzanıyor. Bıyıklar genellikle verinin yayılımının bir ölçüsünü gösterir ve aykırı değerler hariç verinin büyük bir kısmını kapsar. Bu grafikte, en genç yolcunun yaklaşık 0 (bebek) ve en yaşlı yolcunun yaklaşık 64 yaşında olduğu söylenebilir (aykırı değerler hariç).

Aykırı Değerler (Outliers): Bıyıkların üzerinde görünen açık daireler, potansiyel aykırı değerleri temsil eder. Bu grafikte, yaklaşık 65, 70, 74 ve 80 yaşlarında birkaç yolcu olduğu görülüyor. Bu yaşlar, genel dağılımın dışında kalan ve dikkat çekebilecek değerlerdir. Ancak, bu değerlerin veri setindeki gerçek yolculara ait olduğunu unutmamak gerekir.

Dağılımın Şekli: Medyanın kutunun tam ortasında olmaması (alt çeyreğe daha yakın olması) ve üst bıyığın alt bıyığa göre daha uzun olması, yaş dağılımının biraz sağa çarpık (pozitif çarpık) olduğunu düşündürebilir. Bu, daha genç yolcuların sayısının daha fazla olduğu ve yaşlı yolculara doğru kuyruğun uzadığı anlamına gelir.

Genel Yorum: Grafik, Titanic yolcularının yaşlarının genellikle genç ve orta yaş aralığında yoğunlaştığını göstermektedir. Medyan yaş 28 civarındadır ve yolcuların yarısı 20 ile 38 yaş arasındadır. Ancak, özellikle ileri yaşlarda birkaç aykırı değer de bulunmaktadır. En genç yolcular bebeklik çağındadır.


```python
plt.figure(figsize = (8,6))
sns.boxplot(data=fares, color='plum')
plt.title('Fare Distribution of Passengers (Box Graph)')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

## Kategorik Değişkenlerin Sayıları ve Oranları

### Sütun Grafiği

#### bar()

- Matplotlib kütüphanesinin en çok kullanılan sütun (çubuk) grafiği fonksiyonudur.
- Farklı kategorik grupların nicel değerlerini görselleştirmek için kullanılır.
- x ve y değerlerini hesaplayarak vermemiz gerekir. (value_counts())

```python
pclass_count = dataset['Pclass'].value_counts().sort_index() #yolcu sınıflarına göre yolcu sayılarını hesaplar

plt.figure(figsize = (8,6))
plt.bar(pclass_count.index, pclass_count.values, color=['thistle', 'orchid', 'palevioletred'])
plt.title('Number of passengers in passenger classes')
plt.xlabel("Pclass (1 = 1. class, 2 = 2. class, 3 = 3. class)")
plt.ylabel("Passengers count")
plt.xticks(pclass_count.index) 
plt.show()
```

#### countplot()

- Seaborn kütüphanesinden bir fonksiyondur. 
- Kategorik bir sütundaki benzersiz her değerin sayısını sayar ve bunları sütun (çubuk) grafik olarak gösterir.
- value_counts() gibi herhangi bir ek hesaplama yapmadan, kategorik bir değişkenin dağılımını doğrudan görselleştirir.

# İki Değişken Arasındaki İlişkiler

## Sayısal - Sayısal İlişkiler

### Dağılım Grafiği (Scatter Plot)

- Matplotlib -> scatter()
- Seaborn -> scatterplot()

Bu grafik, bir değişkenin diğerine göre nasıl değiştiğini gözlemlemek için çok kullanışlıdır. Grafikteki her bir nokta, veri setindeki tek bir gözlemi temsil eder ve noktanın x eksenindeki konumu bir değişkenin değerini, y eksenindeki konumu ise diğer değişkenin değerini gösterir.


Seaborn kütüphanesinde scatterplot() fonksyionu ile üçüncü bir değişken katıp karşılaştırma yapmamız da mümkün olmaktadır.

1. İlişkinin Yönü (Direction of Relationship):

- Pozitif Korelasyon: Noktalar genel olarak sol alttan sağ üste doğru bir eğilim gösteriyorsa, iki değişken arasında pozitif bir ilişki vardır. Bu, bir değişkenin değeri arttıkça diğerinin de artma eğiliminde olduğu anlamına gelir.
- Negatif Korelasyon: Noktalar genel olarak sol üstten sağ alta doğru bir eğilim gösteriyorsa, iki değişken arasında negatif bir ilişki vardır. Bu, bir değişkenin değeri arttıkça diğerinin azalma eğiliminde olduğu anlamına gelir.
- Korelasyon Yok (No Correlation): Noktalar dağınık bir bulut gibi görünüyorsa ve belirli bir eğilim göstermiyorsa, değişkenler arasında doğrusal bir ilişki olmayabilir.

2. İlişkinin Gücü (Strength of Relationship):

- Güçlü Korelasyon: Noktalar bir çizgiye (doğruya) ne kadar yakınsa, ilişki o kadar güçlüdür.
- Zayıf Korelasyon: Noktalar bir çizginin etrafında dağınık bir şekilde yayılmışsa, ilişki zayıftır.

3. İlişkinin Şekli (Form of Relationship):

- Doğrusal (Linear): Noktalar düz bir çizgi etrafında toplanmışsa.
- Doğrusal Olmayan (Non-linear): Noktalar kavisli bir çizgi (örneğin, parabolik, logaritmik vb.) etrafında toplanmışsa. Bazı durumlarda hiçbir belirgin şekil olmayabilir.

4. Kümelenmeler ve Yoğunluk (Clusters and Density):

- Grafiğin belirli bölgelerinde noktaların daha yoğun olup olmadığını gözlemleyin. Bu, veri setinde belirli özelliklere sahip gözlemlerin kümelendiği alanları gösterebilir. 
- Birden fazla küme varsa, bu, farklı grupların farklı ilişkilere sahip olabileceğini düşündürebilir.

5. Aykırı Değerler (Outliers):

Genel nokta deseninden uzakta duran tekil noktalar olup olmadığına bakın. Bu aykırı değerler, veri girişi hataları, ölçüm hataları veya veri setinde gerçekten istisnai gözlemler olabilir. Bunların nedenini araştırmak faydalı olabilir.

6. Eksen Aralıkları ve Ölçekler:

Eksenlerin aralıklarına ve ölçeklerine dikkat edin. Bazen ölçek seçimi, bir ilişkinin gücünü veya yokluğunu olduğundan farklı gösterebilir.

Dağılım grafiklerinin yorumlamasını "The relationship between age and ticket price (Seaborn)" grafiği üzerinden yapalım.



```python
plt.figure(figsize=(10,7))
plt.scatter(x='Age',
            y='Fare',
            data=dataset,
            alpha=0.6, # Noktaların şeffaflığı (üst üste binen noktaları görmek için)
            s=50, # Noktaların boyutu
            color='mediumaquamarine')
plt.title("The relationship between age and ticket price (matplotlib)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='Age',
            y='Fare',
            data=dataset,
            alpha=0.6, # Noktaların şeffaflığı (üst üste binen noktaları görmek için)
            s=50, # Noktaların boyutu
            color='mediumpurple')
plt.title("The relationship between age and ticket price (Seaborn)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Dağılım grafiklerinin yorumlamasını "The relationship between age and ticket price (Seaborn)" grafiği üzerinden yapalım.

- İlişkinin Yönü ve Gücü: Genellikle belirgin bir doğrusal korelasyon gözlenmez. Yani yaş arttıkça bilet fiyatının düzenli olarak arttığını veya azaldığını gösteren net bir eğilim yoktur. Noktalar genellikle dağınık bir bulut gibi görünür.
- Kümelenmeler/Yoğunluk:
    - Düşük Yaş, Düşük Fiyat: Grafiğin sol alt köşesinde (düşük yaş, düşük bilet fiyatı) büyük bir nokta kümelenmesi ve yoğunlaşma görmeniz muhtemeldir. Bu, 3. sınıf yolcularının (çoğunlukla genç ve uygun fiyatlı) bir göstergesidir.
    - Yüksek Fiyatlar: Y ekseni boyunca yukarı doğru ilerledikçe (daha yüksek bilet fiyatları), nokta sayısının azaldığını, ancak bazı noktaların çok yüksek bilet fiyatlarında yer aldığını göreceksiniz. Bu noktalar genellikle yüksek bilet fiyatları ödeyen 1. sınıf yolcularını temsil eder.
- Aykırı Değerler: Bilet fiyatı ekseninde (Y ekseni) genel dağılımdan çok daha yüksek değerlere sahip birkaç nokta görebilirsiniz. Bunlar çok pahalı biletler almış yolculardır (daha önceki kutu grafikte de görüldüğü gibi). Yaş ekseninde de (X ekseni) çok genç (0-1 yaş) veya çok yaşlı (70+ yaş) aykırı değerler olabilir.
- Sonuç: Yaş ve bilet fiyatı arasında doğrudan güçlü bir ilişki olmamasına rağmen, saçılım grafiği, belirli yaş ve fiyat kombinasyonlarında yolcu yoğunluklarının nasıl dağıldığına dair önemli bilgiler verir. Özellikle düşük yaş ve düşük fiyatlı yolcuların baskınlığını ve yüksek fiyatlı biletlerin genel popülasyon içindeki nadirliğini gösterir.

```python
dataset['Survived_Label'] = dataset['Survived'].map({0: 'Dead', 1: 'Survived'}) 
# Lejantta etiketlerin doğru olması için 0 ve 1 değerlerini map() fonksiyonu ile string değerlere dönüştürdüğümüz yeni bir sütun oluşturduk.
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='Age',
                y='Fare',
                hue='Survived_Label',
                data=dataset,
                alpha=0.7, 
                s=70,
                palette='BuPu')
plt.title("The relationship between age, ticket price and survived (Seaborn)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Life Status') # Zorunlu değil, lejant başlığı tanımlamak için legend() fonksiyonunu kullandık.
plt.show()
```

### Yığılmış Sütun Grafiği

##### Yolcu sınıfı ve hayatta kalma durumunu gösteren yığılmış bir çubuk grafik çizin. Her sınıf için hayatta kalma oranları nasıl farklılaşıyor? 

```python
dataset['Survived_label'] = dataset['Survived'].map({0: 'Dead', 1: 'Survived'})
```

```python
# Her Pclass için Survived durumuna göre sayıları hesaplayın
# Bu, yığılmış çubuk grafik için veri hazırlamanın yaygın bir yoludur.
# size() gruplandırılmış her kombinasyonun sayısını verir
# unstack() ile Survived_Label'ı sütunlara dönüştürerek yığılmış grafik için uygun formatı elde ederiz.
survived_by_pclass = dataset.groupby(['Pclass', 'Survived_label']).size().unstack()
print(survived_by_pclass)
```

```python
# Her sınıf için toplam yolcu sayısı
total_by_pclass = dataset.groupby(['Pclass']).size()
print(total_by_pclass)
```

```python
# Hayatta kalma oranları
survived_rate = survived_by_pclass.div(total_by_pclass, axis=0)
print(survived_rate)
```
